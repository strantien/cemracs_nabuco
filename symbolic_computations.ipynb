```python
import matplotlib.pyplot as plt
import numpy as np

from sympy import symbols, init_printing, pprint, re, im, I, E
from sympy.solvers import solve
from sympy.functions import Abs
from sympy.abc import nu

init_printing()

x, y, nu = symbols('x,y, nu')

solve(x**2+((4-2*y)/(y-3))*x+(y-1)/(y-3),x) #BW
```

## O3

```python
%matplotlib notebook

O3 = solve(-(y*(y+1)*(1-y)/6)+((y*(y+1)*(2-y))/2)*x
+(((y+1)*(1-y)*(2-y)/2)-1)*(x**2)-(y*(1-y)*(2-y)/6)*(x**3),x) #O3

O3

#[float(O3[2].subs(y,u*1e-2)) for u in range(100)]
```

```python
plt.clf()
plt.plot(np.arange(0,1,1e-2),[float(O3[1].subs(y,u*1e-2)) for u in range(100)], label = 'O3[1]')
plt.plot(np.arange(0,1,1e-2),[float(O3[2].subs(y,u*1e-2)) for u in range(100)], label = 'O3[2]')
plt.axis(xmin = 0, ymin = -10, ymax = 1)
plt.grid()
plt.legend()
plt.show()
```

## LWk schemes

```python
LW3 = [-nu*(nu+1)*(1-nu)/6, nu*(nu+1)*(2-nu)/2, 1 -nu*(1+2*nu-nu**2)/2, -nu*(1-nu)*(2-nu)/6]
LW4 = [nu*(nu-1)*(nu+1)*(nu+2)/24, -nu*(nu-2)*(nu+1)*(nu+2)/6, 1 -nu**2*(5-nu**2)/4, -nu*(nu-2)*(nu-1)*(nu+2)/6, nu*(nu-2)*(nu-1)*(nu+1)/24]
LW5 = [nu*(nu-2)*(nu-1)*(nu+1)*(nu+2)/120, -nu*(nu-3)*(nu-1)*(nu+1)*(nu+2)/24, nu*(nu-2)*(nu-3)*(nu+1)*(nu+2)/12, 1 -nu*(nu**4 -3*nu**3 -5*nu**2 +15*nu + 4)/12, nu*(nu-1)*(nu-2)*(nu-3)*(nu+2)/24, -nu*(nu-1)*(nu-2)*(nu-3)*(nu+1)/120]

solve(LW3[0] + LW3[1]*x + (LW3[2]-1)*(x**2) + LW3[3]*(x**3),x)
```

For LW3, exactly same roots as for O3 ...

```python
lw4 = solve(LW4[0] + LW4[1]*x + (LW4[2]-1)*(x**2) + LW4[3]*(x**3) + LW4[4]*(x**4),x)

#mauvaise idee
```

```python
plt.clf()
plt.plot(np.arange(0,1,1e-2),[float(lw4[2].subs(nu,u*1e-2)) for u in range(100)], label = 'lw4[1]')
plt.axis(xmin = 0, ymin = -10, ymax = 1)
plt.grid()
plt.legend()
plt.show()

#mort, valeurs complexes
```
